Задача 1. Урок литературы
Выполните задание, разобранное в уроке.

К уроку литературы нужно было прочитать “Евгения Онегина”. Учитель задаёт вопрос пяти случайным детям. Она задаёт вопрос “Кто написал произведение?” и если ученик не угадывает, то учитель ставит двойку и спрашивает следующего. Если хоть кто-то угадает, то вопросы больше не задаются. Напишите программу, которая посчитает количество двоечников из этих пяти.

```python

to_student = 0
while True:
    book = input('Кто написал войну и мир? - ')
    if book == 'Толстой' or book == 'толстой':
        print('Молодец пять! ')
        break
    print('Неправильно садись два. ')
    to_student += 1
print('Всего двоечников - ', to_student)
```

Задача 2. Начальник
Напишите программу для робота-начальника. Он спрашивает у пользователя, выполнил ли он задание, которые выдавал вчера, и продолжает это делать до тех пор, пока тот не ответит ему “Да, конечно, сделал”.

```python
while True:
    question = input('Ты выполнил задание которое я тебе задал? ')
    if question == 'Да':
        print('Молодец.')
        break
```
